```r
data <- read.xlsx("C:/Users/jongo/Documents/College/Class/GRIDS/Microtelcos/scripts/BC_collapsed_by_group_detailed_for_r_cem.xlsx", sheetIndex = 1)

data <- data %>%
         column_to_rownames('NA.')
```

```r
imbalance(data$has_local, data)
```

```r
todrop <- c('served', 'served25', 'served100', 'served_fib', 
            'comp', 'comp25', 'comp100', 'comp_fib', 
            'max_up', 'max_dn', 'avg_max_up', 'avg_max_dn', 
            'has_coop', 'has_private', 'has_muni', 
            'avg_local', 'avg_coop', 'avg_private', 'avg_muni', 
            'local_telco_under_2000', 'local_telco_2000_4999', 'local_telco_5000_9999', 'local_telco_over_10000', 'isp_count',
            'nointernet', 'has_pc', 'has_pcandbroadband', 'has_internet', 'has_broadband', 'has_mobileonly',
            'avg_d_kbps_wt', 'avg_u_kbps_wt', 'avg_lat_ms_wt')

data_dropped <- data[ , !(names(data) %in% todrop)]
#data_dropped <- data_dropped.frame(na.omit(data_dropped))

data_dropped <- (na.omit(data_dropped))
```

```r
data_dropped
```

```r
#A00001_001    Total Population
#A00002_002    Population Density (per square mile)
#A00002_003    Area (Land, square miles)
#A10003_001    Average Household Size
#A14006_001    Median Household income
#A14008_001    Average Household Income
#A10036_001    Median House Value for All Owner-Occupied Housing Units
```

```r
# m0
match_obj_0.out <- matchit(has_local ~ 
                 A00002_002 + 
                 A14006_001 + 
                 bachelor_more + 
                 tri,
                 data = data_dropped, method = "cem", 
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_0.out

matches_0 <- match.data(match_obj_0.out, include.s.weights = TRUE)

# m1
match_obj_1.out <- matchit(has_local ~ 
                 urban_index + 
                 A14006_001 + 
                 white + hispanic + asian + 
                 bachelor_more +
                 tri,
                 data = data_dropped, method = "cem", 
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_1.out

matches_1 <- match.data(match_obj_1.out, include.s.weights = TRUE)

# m2
match_obj_2.out <- matchit(has_local ~ 
                 urban_index + A00002_002+ 
                 A14006_001 + 
                 white + black + hispanic + asian +
                 tri,
                 data = data_dropped, method = "cem", 
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_2.out

matches_2 <- match.data(match_obj_2.out, include.s.weights = TRUE)

# m3
match_obj_3.out <- matchit(has_local ~ 
                 urban_index +
                 A14006_001 + 
                 white + black + hispanic +
                 unemployed + 
                 tri,
                 data = data_dropped, method = "cem",
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_3.out

matches_3 <- match.data(match_obj_3.out, include.s.weights = TRUE)

# m4
match_obj_4.out <- matchit(has_local ~ 
                 urban_index +
                 A14006_001 + 
                 white + black + hispanic + asian +
                 tri,
                 data = data_dropped, method = "cem", 
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_4.out

matches_4 <- match.data(match_obj_4.out, include.s.weights = TRUE)

# m5
match_obj_5.out <- matchit(has_local ~ 
                 urban_index + 
                 A14006_001 + 
                 bachelor_more + 
                 tri,
                 data = data_dropped, method = "cem", 
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_5.out

matches_5 <- match.data(match_obj_5.out, include.s.weights = TRUE)

# m6
match_obj_6.out <- matchit(has_local ~ 
                 urban_index + 
                 A00002_002 +
                 A14006_001 + 
                 bachelor_more + 
                 tri,
                 data = data_dropped, method = "cem", 
                 s.weights= ~A00001_001, include.s.weights = TRUE)

match_obj_6.out

matches_6 <- match.data(match_obj_6.out, include.s.weights = TRUE)
```
